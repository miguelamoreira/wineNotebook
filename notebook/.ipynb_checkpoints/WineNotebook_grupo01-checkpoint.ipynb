- Como se trata de um valor reduzido em compara√ß√£o aos restantes valores, podemos elimin√°-los sem prejudicar o resto da an√°lise de dados e a cria√ß√£o dos modelos de classifica√ß√£o e de regress√£o

```python
# fun√ß√£o para remover os outliers de cada atributo
def remove_outliers(df):
    df_clean = df.copy()
    for col in df.select_dtypes(include=[np.number]).columns:
        mean = df[col].mean()
        std = df[col].std()
        cutoff = std * 3
        lower, upper = mean - cutoff, mean + cutoff
        df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]
    return df_clean

df = remove_outliers(df)
```

## Modelos de Classifica√ß√£o
### Objetivo: Desenvolver modelos de classifica√ß√£o que permitam prever corretamente o tipo de vinho.
Uma vez que os valores de qualidade ('quality') de ambos os tipos de vinhos s√£o bastante semelhantes, podemos remover esta coluna antes de proceder √† cria√ß√£o dos modelos.

Para analisar a capacidade de previs√£o dos diferentes modelos criados (Decision Tree Classifier, Random Forest Classifier e SVC), v√£o ser avaliados os seguintes par√¢metros:
- Accuracy: mede a propor√ß√£o de previs√µes feitas corretamente em compara√ß√£o ao n√∫mero total de previs√µes. Os valores variam entre 0 e 1. Quanto mais pr√≥ximo de 1, onde 1 representa uma taxa de precis√£o ideal (100%);
- F1-Score: representa a m√©dia entre a precis√£o (precision) e recall. Este par√¢metro √© importante na an√°lise de modelos com base no dataset a ser utilizado uma vez que toma em considera√ß√£o desbalanceamentos de dados que possam ocorrer. Durante a fase de an√°lise de dados, foi evidenciado que o dataset continha um n√∫mero bastante superior de informa√ß√£o relativa aos vinhos tintos do que de vinhos brancos; 
- Precision: mede a propor√ß√£o de verdadeiros positivos em compara√ß√£o ao n√∫mero de todas as amostras previstas como positivas (verdadeiros positivos e falsos positivos);
- Recall: mede a propor√ß√£o de verdadeiros positivos em compara√ß√£o ao n√∫mero de todas as amostras realmente positivas (verdadeiros positivos e falsos negativos).

```python
# defini√ß√£o das vari√°veis X (atributos) e y (tipo de vinho)
X = df.drop(columns=['quality', 'type'], axis=1)
y = df['type']
```

```python
# divis√£o dos valores a serem utilizados para teste e treino
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)
```

```python
# fun√ß√£o para avaliar accuracy, f1-score, precision e recall
def evaluateClassification (reais, previstos):
    _accuracy = accuracy_score(reais, previstos)
    _f1 = f1_score(reais, previstos, average='weighted')
    _precision = precision_score(reais, previstos, average='weighted')
    _recall = recall_score(reais, previstos, average='weighted')
    return round(_accuracy,2), round(_f1,2), round(_precision,2), round(_recall,2)

# fun√ß√£o para imprimir os valores obtidos na fun√ß√£o evaluateClassification
def print_evaluateClassification(model, reais, previstos):
    accuracy, f1, precision, recall = evaluateClassification(reais, previstos)
    print("------------------------------------------")
    print("Accuracy: ", accuracy)
    print("F1-Score: ", f1)
    print("Precision: ", precision)
    print("Recall: ", recall)
    print("------------------------------------------")
    
    results=np.array([model, accuracy, f1, precision, recall])
    results_classification.loc[len(results_classification)+1]= results

# cria√ß√£o da tabela para armazenar os valores obtidos 
results_classification = pd.DataFrame(columns=['Model', 'Accuracy','F1','Precision','Recall'])
```

## - Modelo de Classifica√ß√£o 1: Decision Tree Classifier

```python
# cria√ß√£o do modelo de classifica√ß√£o Decision Trees (DTC)
dtc = tree.DecisionTreeClassifier(criterion='entropy', max_depth=5, random_state=24)

# treino do modelo DTC
dtc = dtc.fit(X_train, y_train)
```

```python
# previs√£o do modelo DTC
dtc_pred = dtc.predict(X_test)
```

```python
# avalia√ß√£o de accuracy, f1-score, precision e recall do modelo DTC
print('Decision Tree Classifier')
print_evaluateClassification('Decision Tree Classifier', y_test, dtc_pred)
```

## Modelos de Regress√£o
### Objetivo: Desenvolver modelos de regress√£o que permitam prever a qualidade do vinho.
Uma vez que o atributo 'quality' √© importante para esta an√°lise, √© necess√°rio fazer uma nova defini√ß√£o dos valores atribu√≠dos √†s vari√°veis X e y

Para analisar a capacidade de previs√£o dos diferentes modelos criados (Linear Regression, Random Forest Regressor e XGBoost), v√£o ser avaliados os seguintes par√¢metros:
- MAE (Mean Absolute Error): representa a m√©dia dos erros absolutos entre os valores previstos e os valores reais. Quanto menor o MAE, melhor o desempenho do modelo;
- MSE (Mean Square Error): representa a m√©dia dos erros elevados ao quadrado entre os valores previstos e os valores reais. O MSE penaliza os erros mais severamente que o MAE devido a elev√°-los ao quadrado;
- RMSE (Root Mean Squared Error): representa a raiz da quadrado do MSE, aproximando os valores dos erros obtidos da escala dos valores reais;
- R2 (Coefficient of Determination): mede a propor√ß√£o da variabilidade nos dados de resposta dados pelos modelos, podendo este valor variar entre 0 e 1. Quanto mais pr√≥ximo de 1, melhor √© a capacidade de previs√£o do modelo.

```python
# defini√ß√£o das vari√°veis X (atributos) e y (tipo de vinho)
X = df.drop(columns=['quality'], axis=1)
y = df['quality']
```

```python
# divis√£o dos valores a serem utilizados para teste e treino
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24)
```

```python
# fun√ß√£o para avaliar os erros (MAE, MSE, RMSE, R2) de cada modelo
def evaluate (reais, previstos):
    _mae=mae(reais, previstos)
    _mse=mse(reais, previstos)
    _rmse=np.sqrt(_mse)
    _r2=r2_score(reais, previstos)
    return round(_mae,2), round(_mse,2), round(_rmse,2), round(_r2,2)

# fun√ß√£o para imprimir os valores obtidos na fun√ß√£o de avalia√ß√£o 
def print_evaluateRegression(model, reais, previstos):
    mae, mse, rmse, r2 = evaluate(reais,previstos)
    print("------------------------------------------")
    print("MAE:", mae)
    print("MSE:", mse)
    print("RMSE:", rmse)
    print("R2 score:", r2)
    print("------------------------------------------")
    
    results=np.array([model,mae,mse,rmse,r2])
    results_regression.loc[len(results_regression)+1]= results

# cria√ß√£o da tabela para armazenar os valores obtidos 
results_regression = pd.DataFrame(columns=['Model', 'MAE','MSE','RMSE','R2'])
```

## - Modelo de Regress√£o 1: Linear Regression

```python
# cria√ß√£o do modelo Linear Regression (LR)
lr = LinearRegression()

# treino do modelo criado
lr.fit(X_train, y_train)
```

```python
# previs√£o do modelo LR
lr_pred = lr.predict(X_test)
```

```python
# avalia√ß√£o de MAE, MSE, RMSE e R2 do modelo LR
print('Linear Regression')
print_evaluateRegression('Linear Regression', y_test, lr_pred)
```

### Interpreta√ß√£o dos resultados obtidos:
- Os valores de MAE, MSE e RMSE nem s√£o muito elevados nem muito baixos, sendo o valor mais alto de 0.69 referente a RMSE. Isto faz com que o modelo de Linear Regression apresente um grau consider√°vel de previs√£o de erro, podendo existir outros modelos mais adequados ao que √© pretendido;
